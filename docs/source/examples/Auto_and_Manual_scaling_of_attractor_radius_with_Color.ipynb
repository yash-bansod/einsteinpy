# _Static Geodesic Plotter class_

### Importing the required modules

```python
import numpy as np
import astropy.units as u
from einsteinpy.plotting import StaticGeodesicPlotter
from einsteinpy.metric import Schwarzschild
from einsteinpy.coordinates import CartesianDifferential, SphericalDifferential
```

### Plotting Earth around Sun (Auto scaling is used) 

#### Since the Schwarzchild Radius of the Sun is small, it won't be visible in the plot. That is why Auto Scaling is used.
#### Auto scaling is used when the scaling parameters are not passed.

```python
M = 1.989e30 * u.kg  # mass of sun
distance = 147.09e6 * u.km
speed_at_perihelion = 30.29 * u.km / u.s
omega = (u.rad * speed_at_perihelion) / distance

sph_obj = SphericalDifferential(distance, np.pi / 2 * u.rad, np.pi * u.rad,
                               0 * u.km / u.s, 0 * u.rad / u.s, omega)

# Set lambda to complete an year.
# Lambda is always specified in secs
end_lambda = ((1 * u.year).to(u.s)).value
# Choosing stepsize for ODE solver to be 5 minutes
stepsize = ((5 * u.min).to(u.s)).value

sgp = StaticGeodesicPlotter(M)
sgp.plot(sph_obj, end_lambda, stepsize)
sgp.show()
```

### Defining various parameters
- Mass of the attractor(M)
- Initial position and velocity vectors of test partcle

```python
M = 6e24 * u.kg
sph_obj = SphericalDifferential(130*u.m, np.pi/2*u.rad, -np.pi/8*u.rad,
                                0*u.m/u.s, 0*u.rad/u.s, 1900*u.rad/u.s)
```

### Using Auto scaling

```python
"""
Without color and Without Manual Scaling.
Not giving scaling factor implies auto scaling in which radius is scaled to minimum of the following: 
1) 1/12 the minimum of the length of X and Y axix.
2) Minimum distance between attractor and the object revolving around it.
"""
%matplotlib inline
obj = StaticGeodesicPlotter(M)
obj.plot(sph_obj, 0.002, 5e-8)
obj.show()
```

### Get the radius of the attractor as shown in the plot

```python
"""
Getting the current plotted radius of the attractor with the Auto scaling done above 
""" 
print(obj.get_curr_plot_radius)
```

### Using color and Manual Scaling

```python
%matplotlib inline
obj = StaticGeodesicPlotter(M, attractor_color = "red", attractor_radius_scale = 1500)
obj.plot(sph_obj, 0.002, 5e-8)
obj.show()
```

### Get the radius of the attractor as shown in the plot

```python
"""
Getting the current plotted radius of the attractor with the Manual scaling done above 
""" 
print(obj.get_curr_plot_radius)
```

# _Scatter Geodesic Plotter class_

### Importing the required modules

```python
from einsteinpy.plotting import ScatterGeodesicPlotter
```

### Defining various parameters
- Mass of the attractor(M)
- Initial position and velocity vectors of test partcle

```python
M = 6e24 * u.kg
sph_obj = SphericalDifferential(130*u.m, np.pi/2*u.rad, -np.pi/8*u.rad,
                                0*u.m/u.s, 0*u.rad/u.s, 1900*u.rad/u.s)
```

### Without Color

```python
%matplotlib inline
obj = ScatterGeodesicPlotter(M)
obj.plot(sph_obj, 0.002, 5e-8)
obj.show()
```

### With Color

```python
%matplotlib inline
obj = ScatterGeodesicPlotter(M, attractor_color = "blue", cmap_color = "Reds")
obj.plot(sph_obj, 0.002, 5e-8)
obj.show()
```
